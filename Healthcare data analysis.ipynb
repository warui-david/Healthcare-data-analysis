### For Each District Find Out The Most Common Surgery Type That Is Being Performed.

```python
for district in df['DISTRICT_NAME'].unique():
    district_data = df[df['DISTRICT_NAME'] == district]['SURGERY'].value_counts().head(1)
    most_frequent_disease = district_data.index[0]  # Get the first index (disease name)
    count = district_data.iloc[0]  # Get the count of the most frequent disease
    print(f"District: {district}\nDisease and Count: {most_frequent_disease}\nCount: {count}\n")
```

Our analysis shows that Nellore performs Maintenance Hemodialysis For Crf which has the highest surgery count.

We can also notice that there are two surgeries involved in all districts, e.i. Hemodialysis and Longbone Fracture.

### For each district, find the average claim amount

```python
for i in df['DISTRICT_NAME'].unique():
    print("District: {}\n Average Claim Amount: {}\n '============'\n".format(i,df[df['DISTRICT_NAME']==i]['CLAIM_AMOUNT'].mean()))    
```

### Group Summary Statistics

```python
result = df.groupby('CATEGORY_NAME')[['PREAUTH_AMT', 'AGE', 'CLAIM_AMOUNT']].mean().sort_values(by='PREAUTH_AMT', ascending=False)
result
```

#### Cochlear Implant Surgery seems to be the costliest surgery, costing ₹520000 on average. Prostheses costs ₹1200, the cheapest. The youngest age group also happens to be for Cochlear Implant Surgery: 1.58 years, while Neurology has an average patient age of ~56.

### Summary statistics for each hospital

```python
result2 = df.groupby('HOSP_NAME')[['PREAUTH_AMT', 'AGE', 'CLAIM_AMOUNT']].mean().sort_values(by='PREAUTH_AMT', ascending=False)
result2
```

#### Maa Hospitals Pvt. Ltd. cost the highest with pre-Authorised amount of 520000.

### Most common surgery and count per age group

```python
for i in sorted(df['AGE'].unique()):
    most_common_category = df[df['AGE'] == i]['CATEGORY_NAME'].value_counts().idxmax()
    count = df[df['AGE'] == i]['CATEGORY_NAME'].value_counts().max()
    print("Age Group: {}\nMost Common Surgery: {}\nCount: {}\n'============'\n".format(i, most_common_category, count))
```

### Pre-Authorised Amount Over-time
We will use the average of the last 12 data points to check the trend. We will use a technique called rolling average.

```python
preauthorised=df['PREAUTH_AMT']

preauthorised.rolling(12).mean().plot(figsize=(20,10),linewidth=2,fontsize=20)

plt.xlabel('Year', fontsize=20)
```

#### The highest pre-authorised amounts are seen in the month of July 2015 and mid 2016.

#### We can also notice that there has been a lot of clustered pre-authorised amounts between 25000 and 100000 in January 2016 and July 2017.

### Claim amount over time

```python
preauthorised=df['CLAIM_AMOUNT']

preauthorised.rolling(12).mean().plot(figsize=(20,10),linewidth=2,fontsize=20)

plt.xlabel('Year', fontsize=20)
```
